```python
tags_set = []
db = db_client["tweets_pipeline_v2"].find({"$or":[{"status": "graphicone_search"},
                                                  {"status": "graphicone_feed"},
                                                  {"status": "validated"},
                                                 ],
                                          }, no_cursor_timeout=True)
for i in tqdm(db, total = db.count()):
    if "grafeed_confirmed" in i:
        tags_set += i["grafeed_confirmed"]
    else:
        if "confirmed_after_validate" in i:
            tags_set += i["confirmed_after_validate"]
```
