# GETTING ALL IMGs

```python
def GetAllImages(link, args):
    pageContent = requests.get(link).content 
    soup = bs(pageContent, "lxml")
    Images = []
    for i in [i.get('href') for i in soup.find_all("a")]:
        basename = os.path.basename(i)
        extension = basename.rpartition('.')[-1]
        if extension in args["Formats"]:
            ImgLink = 'https://www.sec.gov' + i
            Images.append(ImgLink)
    return Images
```

```python
index = 23
linkExample = npDB[:,-1][index]
print(linkExample)
images = GetAllImages(linkExample, args)
images
```
