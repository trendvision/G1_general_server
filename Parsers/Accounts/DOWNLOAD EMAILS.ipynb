```python
import pickle
import pandas as pd
import re
import os
```

```python
bbgvisualdata = pickle.load( open( "bbgvisualdata.p", "rb" ) )
WSJGraphics = pickle.load( open( "WSJGraphics.p", "rb" ) )
```

```python
data = WSJGraphics + bbgvisualdata
```

```python
len(data)
```

```python
def find_email(text):
    regex = r'([\w0-9._-]+@[\w0-9._-]+\.[\w0-9_-]+)'
    return re.findall(regex, text, re.M|re.I)
```

```python
clean_data = []
for user in data:
    email = None
    if find_email(user.description):
        email = find_email(user.description)
    expanded_url = None    
    if "url" in user.entities:
        expanded_url = user.entities["url"]['urls'][0]["expanded_url"]        
        if expanded_url is not None and find_email(expanded_url):
            email = find_email(expanded_url)
                            
    clean_data.append([user.name, expanded_url, email])
```

```python
len(clean_data)
```

```python
df = pd.DataFrame(clean_data, columns = ['name', 'website', 'email'])
df
```

```python
def websiteCleaner(website):    
    website = re.sub("http://", "", website)
    website = re.sub("https://", "", website)
    website = re.sub("www.", "", website)
    return website
```

```python
def findPerson(website, fullname):
    if website is not None:
        website = website.lower()    
        if len(fullname.split()) > 1 and blackListForWebsites(website):
            website = websiteCleaner(website)
            link = os.path.join("https://hunter.io/find/",website,'%20'.join(fullname.split()))
            return link
        else:
            return None
    else:
        return None
```

```python
def blackListForWebsites(website):
    if not re.search(r"instagram|medium|patreon|behance|twitter", website):
        return True
```

```python
df['link'] = df.apply(lambda x: findPerson(x["website"], x["name"]), axis = 1)
```

```python
accounts_clean_data = df[(~pd.isnull(df.email)) | (~pd.isnull(df.link))]
accounts_clean_data
```

```python
accounts_clean_data.to_csv("accounts_clean_data.csv", index=False)
```
