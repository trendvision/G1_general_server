```python
import bs4 as bs
```

```python
ua = UserAgent()
```

```python
r = requests.get("https://finviz.com/quote.ashx?t=AMZN", headers={'User-Agent': ua.random})
r
```

```python
soup = bs.BeautifulSoup(r.text,'lxml')
soup
```

```python
data = []
table = soup.find('table', attrs={'class':"fullview-ratings-outer"})
# print(table)
# table_body = table.find('tbody')
# print(table_body)
rows = table.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values
```

```python
import pandas as pd
```

```python
pd.DataFrame(data).to_excel("finviz_example.xlsx")
```
