```python
import requests
import time
import re
from collections import Counter
import pandas as pd
from bs4 import BeautifulSoup as bs
import os
import json
from tqdm import tqdm
tqdm.pandas()
from itertools import chain
from user_agent import generate_user_agent
```

```python
def request_links_Session(url, attempts):
    for _ in range(0, attempts):
        try:
            session = requests.Session()
            headers =  {'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                        'User-Agent': str(generate_user_agent(device_type='desktop', os=('mac', 'linux'))),
                        'Connection': 'Keep-Alive',
                        'Accept-Encoding': 'gzip, deflate',
                        'Accept-Language': "en-US,en;q=0.8"}
            page = session.get(url, headers=headers, timeout = 6)
            content = bs(page.content, features="lxml")
            return content            
        except Exception as err:
            print(err)
```

```python
def get_all_articles_links(content):
    contents = ["https://seekingalpha.com"+i['href'] for i in content.find_all("a", class_="a-title")]
    return contents
```

```python
def get_all_article_images_links(content):
    contents = [img.get('src') for img in content.find_all('img')]
    return contents        
```

```python
####################### DOWNLOADING #############################

def download(url, name, Folder, SubFolder):
    extension = url[-4:]
    r = requests.get(url, timeout=5.0) 
    if extension in [".jpg", ".png"]:        
        directory = os.path.join(Folder, SubFolder)
        if not os.path.exists(directory):
            os.makedirs(directory)
        path = os.path.join(Folder, SubFolder, (str(name) + extension))
        with open(path, 'wb') as f:
            f.write(r.content)       
            
def DownloadLinks(Links, Folder, SubFolder):
    counter=0
    for url in tqdm(Links, postfix=""):
        counter += 1
        try:
            name = counter            
            download(url, name, Folder, SubFolder)
        except Exception as err:
            print(err)
```

```python
images_links = []
# base_url = "https://seekingalpha.com/stock-ideas/long-ideas"
base_url = "https://seekingalpha.com/stock-ideas/technology"
for i in range(100):
    content = request_links_Session(base_url+"?page="+str(i), 2)
    time.sleep(3)
    print("Page:", i)
    page_contents = get_all_articles_links(content)
    for article, page_content_link in enumerate(page_contents):
        print("Article:",article)
        page_content = request_links_Session(page_content_link, 2)        
        images_links += get_all_article_images_links(page_content)
        time.sleep(2)
    time.sleep(3)
```
